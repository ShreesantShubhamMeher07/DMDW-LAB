```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data =pd.read_csv('https://raw.githubusercontent.com/archana1822/DMDW-Lab/main/Toyota.csv')
```

```python
data.head()
```

# LINE PLOT

```python
plt.plot(data['MetColor'],data['HP'])
plt.title('MetColor Vs HP')
plt.xlabel('MetColor')
plt.ylabel('HP')
plt.show()
```

# BAR PLOT

```python
plt.bar(data['KM'],data['Weight'],color='Green')

plt.xlabel('KM')
plt.ylabel('Weight')
plt.show()
```

# HISTOGRAM

```python
plt.hist(data['Weight'],color='Green')
plt.ylabel('Weight')
plt.show()
```

# SCATTER PLOT

```python
plt.scatter(data['CC'],data['Weight'],color='Red')
plt.ylabel('Weight')
plt.xlabel('CC')
plt.show()
```

# PIE CHART

```python
cars = ['AUDI', 'BMW', 'FORD', 
        'TESLA', 'JAGUAR', 'MERCEDES'] 
  
data = [23, 17, 35, 29, 12, 41] 
  

fig = plt.figure(figsize =(10, 7)) 
plt.pie(data, labels = cars) 

plt.show() 
```

# AREA UNDER CURVE

```python
sns.distplot( [data['Age']] )
```

# BOX PLOT

```python

sns.boxplot( y=data["Price"] );
```

# HEAT MAP

```python
data_corr=data.corr()
sns.heatmap(data_corr,annot=True,linecolor='black',linewidths=1)
```
